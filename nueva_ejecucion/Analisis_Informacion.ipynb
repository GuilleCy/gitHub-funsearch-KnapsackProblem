# aplicacion de solver

```python
import pandas as pd
from my_greedy_heuristic import MyGreedyHeuristic
```

```python
df.rename(columns={'muestra_id': 'lote_id'}, inplace=True)
```

```python
df_recuperado=df.copy()
```

```python
df_recuperado.rename(columns={'muestra_id': 'lote_id'}, inplace=True)
```

```python
df_recuperado.head()
```

```python
#============================================================
# 2Ô∏è‚É£ Aplicar heur√≠stica MyGreedyHeuristic a cada instancia
# ============================================================
resultados = []

for idx, inst in df_recuperado.iterrows():
    try:
        weights = inst["pesos"]
        values = inst["valores"]
        capacity = inst["capacidad"]

        # Instanciar la heur√≠stica (usa el constructor de la clase)
        solver = MyGreedyHeuristic(weights, values, capacity)

        # Resolver el problema usando la heur√≠stica definida
        resultado = solver.solve()

        resultados.append({
            "lote_id": inst["lote_id"],
            "valor_total_fun": resultado["total_value"],
            "peso_usado_fun": resultado["total_peso_usado"],
            "tiempo_fun": resultado["solve_time"]
        })

        print(f"‚úÖ Lote {inst['lote_id']} resuelto correctamente")

    except Exception as e:
        print(f"‚ùå Error en lote {inst['lote_id']}: {e}")
```

```python
df_resultados = pd.DataFrame(resultados)
```

```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ============================================================
# N√∫mero de simulaciones
# ============================================================
N = 100

# Lista para acumular resultados de variaci√≥n
resultados = []

for i in range(N):
    print(f"\n===== Simulaci√≥n {i+1} de {N} =====")

    # === Generar muestras ===
    generador = GeneradorMuestrasMochila()
    df_muestras = generador.crear_muestras()
    df1 = df_muestras[['muestra_id', 'num_items', 'capacidad', 'totalPeso']].copy()
    df1 = df1.rename(columns={'muestra_id': 'lote_id'})

    # === Resolver con OR-Tools ===
    evaluador = EvaluadorMochila(df_muestras, hash_id_func=generador.get_hash_id)
    evaluador.resolver_muestras()
    df2 = df_muestras[['muestra_id', 'valor_total', 'peso_total']].copy()
    df2 = df2.rename(columns={'muestra_id': 'lote_id',
                              'valor_total': 'ORTools',
                              'peso_total': 'peso_ORTools'})

    # === Resolver con MILP (heur√≠stica FunSearch) ===
    resultados_fun = []
    for idx, inst in df_muestras.iterrows():
        try:
            weights = inst["pesos"]
            values = inst["valores"]
            capacity = inst["capacidad"]

            solver = MyGreedyHeuristic(weights, values, capacity)
            resultado = solver.solve()

            resultados_fun.append({
                "lote_id": inst["muestra_id"],
                "MLIP-Greedy": resultado["total_value"],
                "peso_MLIP": resultado["total_peso_usado"],
                "tiempo_fun": resultado["solve_time"]
            })

        except Exception as e:
            print(f"‚ùå Error en lote {inst['muestra_id']}: {e}")

    df3 = pd.DataFrame(resultados_fun)

    # === Unir resultados ===
    df_12 = pd.merge(df1, df2, on="lote_id", how="inner")
    df_final = pd.merge(df_12, df3, on="lote_id", how="inner")

    # === Calcular variaci√≥n porcentual ===
    df_final["variacion_MLIP_ORTOOLS"] = (
        (df_final["MLIP-Greedy"] - df_final["ORTools"]) / df_final["ORTools"]
    ) * 100

    resultados.append(df_final[["lote_id", "variacion_MLIP_ORTOOLS"]])

# ============================================================
# Combinar resultados de todas las simulaciones
# ============================================================
df_resultados = pd.concat(resultados, ignore_index=True)

# Agrupar por muestra (lote_id) para obtener media y desviaci√≥n
df_stats = df_resultados.groupby("lote_id")["variacion_MLIP_ORTOOLS"].agg(["mean", "std"]).reset_index()

# ============================================================
# Graficar resultados
# ============================================================
plt.figure(figsize=(12, 6))
plt.plot(df_stats["lote_id"], df_stats["mean"], color="blue", marker="o", label="Variaci√≥n media (%)")
plt.fill_between(df_stats["lote_id"],
                 df_stats["mean"] - df_stats["std"],
                 df_stats["mean"] + df_stats["std"],
                 color="blue", alpha=0.2, label="¬±1 Desv. est√°ndar")

plt.axhline(0, color="gray", linestyle="--", linewidth=1)
plt.xlabel("Lote ID")
plt.ylabel("Variaci√≥n (%)")
plt.title(f"Comparaci√≥n MILP-Greedy vs OR-Tools ({N} simulaciones)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
